## Instance method
Q1. Write a Python program to display a student's name.

```python
# Q1. Write a Python program to display a student's name using an instance method.
class Student:
    def __init__(self, name):
        self.name = name

    def show_name(self):
        print("Student Name:", self.name)

s1 = Student("Priya")
s1.show_name()
```

Q2. Write a program to display an employee's salary.

```python
class Employee:
    def __init__(self, salary):
        self.salary = salary

    def show_salary(self):
        print("Salary:", self.salary)

e1 = Employee(50000)
e1.show_salary()
```

Q3. Write a program to calculate the area of a rectangle.

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        print("Area:", self.length * self.width)

r = Rectangle(10, 5)
r.area()
```

Class Method

Q1. Write a program to display the school name.

```python
class School:
    school_name = "ABC School"

    @classmethod
    def show_school(cls):
        print(cls.school_name)

School.show_school()
```

Q2. Write a program to change the company name.

```python
class Company:
    company = "Infosys"

    @classmethod
    def change_company(cls, name):
        cls.company = name

Company.change_company("TCS")
print(Company.company)
```

Static Method

Q1. Write a program to add two numbers

```python
class Calculator:
    @staticmethod
    def add(a, b):
        print("Sum:", a + b)

Calculator.add(10, 20)
```

Q2. Write a program to find the square of a number.

```python
class Math:
    @staticmethod
    def square(n):
        print("Square:", n * n)

Math.square(6)
```

Q3. Write a program to check whether a number is even or odd.

```python
class Number:
    @staticmethod
    def check(n):
        if n % 2 == 0:
            print("Even")
        else:
            print("Odd")

Number.check(7)
```

Q4. Write a program to multiply two numbers.

```python
class Multiply:
    @staticmethod
    def product(a, b):
        print("Product:", a * b)

Multiply.product(4, 5)
```

Q5. Write a program to convert Celsius to Fahrenheit.

```python
class Temperature:
    @staticmethod
    def convert(c):
        print("Fahrenheit:", (c * 9/5) + 32)

Temperature.convert(25)
```
